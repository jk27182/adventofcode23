```python
from typing import Tuple
```

```python
with open('2023_12_01/data.txt', 'r') as f:
    text_corpus = f.readlines()
text_corpus
```

```python
nums = {str(i) for i in range(10)}
def find_digits(text: str) -> Tuple[int]:
    left = 0
    right = len(text) - 1
    while left <= right:
        left_elem = text[left]
        right_elem = text[right]
        if left_elem in nums and right_elem in nums:
            return int(left_elem + right_elem)
        elif left_elem not in nums:
            left += 1
        elif right_elem not in nums:
            right -= 1
        else:
            left += 1
            right -= 1
```

```python
res = 0
for line in text_corpus:
    text = line.split('\n')[0]
    num = find_digits(text)
    # print(num)
    res += num
```

```python
# Trie implementation
class TrieNode:
    def __init__(self, text: str = '') -> None:
        self.text = text
        self.children = dict()
        self.is_word = False

class PrefixTree:
    def __init__(self) -> None:
        self.root = TrieNode()

    def insert(self, word: str) -> None:
        current = self.root
        for i, char in enumerate(word):
            if char not in current.children:
                prefix = word[0:i+1]
                current.children[char] = TrieNode(prefix)
            current = current.children[char]
        current.is_word = True

    def starts_with(self, text: str) -> None:
        pass

    def find(self, word):
        '''
        Returns the TrieNode representing the given word if it exists
        and None otherwise.
        '''
        current = self.root
        for char in word:
            if char not in current.children:
                return False
            current = current.children[char]

        # New code, None returned implicitly if this is False
        return True
```

```python
# build prefix tree of numbers
trie = PrefixTree()
written_nums = ['one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']
for written_num in written_nums:
    trie.insert(written_num)
```
