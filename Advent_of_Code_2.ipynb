```python
import re

import pandas as pd
```

```python
with open('2023_12_02/data.txt', 'r') as f:
    data_raw = f.readlines()
data_raw
```

```python
df = pd.read_csv('2023_12_02/data.txt', delimiter=';', header=None)
df
```

```python
df[['ID', 0]] =  df.iloc[:, 0].str.split(':', expand=True)
df['ID'] = df['ID'].str.split(' ').str[-1]
df
```

```python
df_long = df.melt(id_vars='ID', value_vars=[0, 1, 2,3, 4, 5], value_name='value').drop('variable', axis=1)
```

```python
df_long[df_long['ID'] == '1']['value'].to_list()
```

```python
test = '1 green, 4 blue'
```

```python
re.findall(r'(\d) (blue|green|red)', test)
```

```python
df_long
```

```python
# Find respective counts for each color with a regex
# If color does not appear (NaN), replace NaN with 0
# 
df_counts_per_color = df_long['value']\
    .str.extractall(r'(\d*) (green|blue|red)')\
    .reset_index()\
    .drop(columns='match')\
    .pivot(index='level_0', columns=1)\
    .fillna(0)\
    .droplevel(level=0, axis=1)\
    [['blue', 'green', 'red']]
df_counts_per_color
```

```python
df_clean = pd.merge(
    df_long,
    df_counts_per_color,
    left_index=True,
    right_index=True
)
df_clean
```

```python
df_clean = df_clean.astype({'ID': int, 'blue': int, 'green': int, 'red': int })
```

```python
# Finally check if games are possible
red_cubes = 12
green_cubes = 13
blue_cubes = 14
df_clean['game1_possible'] = (
    (df_clean['red'] <= red_cubes)
    & (df_clean['green'] <= green_cubes)
    & (df_clean['blue'] <= blue_cubes)
)
df_clean
```

```python
df_possible_games = df_clean.groupby('ID')['game1_possible'].all()
sum(df_possible_games[df_possible_games].index)
```
